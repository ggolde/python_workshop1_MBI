## Example 3.1

Topics: pandas DataFrames, missing values, boolean indexing, creating columns, sorting, aggregation

Create a cleaned leaderboard from the student quiz results in `df`.

Tasks:
- Replace missing values in `score` with the mean of the `score` column.

- Replace missing values in `attempts` with 1.

- Keep only the rows where `hours_studied >= 2`.

- Create a new column `score_per_attempt = score / attempts`.

- Create a new column `passed` where `score >= 70`.

- Sort the result by `score_per_attempt` from highest to lowest.

- Return the cleaned DataFrame, the mean score of students who passed, and the number of students who did not pass.

Tips:
- `fillna()` can replace missing values in a Series.
- `copy()` is useful when you want to avoid modifying the original DataFrame.
- `shape[0]` gives the number of rows in a DataFrame.

```python
# setup
df = pd.DataFrame(
    {
        'student': ['Ana', 'Ben', 'Cleo', 'Darius', 'Eli', 'Fern'],
        'hours_studied': [3.5, 1.0, 2.5, 4.0, 2.0, 5.0],
        'score': [88.0, np.nan, 73.0, 95.0, 64.0, np.nan],
        'attempts': [2, 1, np.nan, 4, 2, 5],
    }
)

# solution
def solution3_1(df: pd.DataFrame) -> tuple[pd.DataFrame, float, int]:
    d = df.copy()
    d['score'] = d['score'].fillna(d['score'].mean())
    d['attempts'] = d['attempts'].fillna(1)

    d = d[d['hours_studied'] >= 2].copy()
    d['score_per_attempt'] = d['score'] / d['attempts']
    d['passed'] = d['score'] >= 70
    d = d.sort_values('score_per_attempt', ascending=False)

    mean_passed_score = d.loc[d['passed'], 'score'].mean()
    num_failed = d.loc[~d['passed']].shape[0]

    return d, round(mean_passed_score, 2), num_failed

leaderboard, mean_passed_score, num_failed = solution3_1(df)

print(leaderboard, '\n')
print(mean_passed_score)
print(num_failed)
```

```python
# method chaining solution

def solution3_1_b(df: pd.DataFrame) -> tuple[pd.DataFrame, float, int]:
    d: pd.DataFrame = (
        df.copy()
        .assign(
            score = lambda d: d['score'].fillna(d['score'].mean()),
            attempts = lambda d: d['attempts'].fillna(1),
            score_per_attempt = lambda d: d['score'] / d['attempts'],
            passed = lambda d: d['score'] >= 70,
        )
        .loc[lambda d: (d['hours_studied'] >= 2)]
        .sort_values('score_per_attempt', ascending=False)
    )
    mean_passed_score = d.loc[d['passed'], 'score'].mean()
    num_failed = d.loc[~d['passed']].shape[0]

    return d, round(mean_passed_score, 2), num_failed

leaderboard, mean_passed_score, num_failed = solution3_1_b(df)

print(leaderboard, '\n')
print(mean_passed_score)
print(num_failed)
```

## Example 3.2

Topics: pandas DataFrames, merging, creating columns, `loc`, groupby, aggregation, rounding

Use the `students` and `scores` DataFrames to build a grouped performance summary.

Tasks:
- Merge the DataFrames on `student_id`.

- Create a new column `average_score` from `exam1` and `exam2`.

- Create a new column `passed` where `average_score >= 70`.

- Use `loc` to create a `letter_grade` column with `A`, `B`, `C`, and `F`.

- Group by `group` and calculate the mean `average_score` and the number of students who passed.

- Round numeric results to 2 decimal places.

Tips:
- `merge()` combines DataFrames using a shared column.
- You can assign values to selected rows with `df.loc[condition, 'column'] = value`.
- `sum()` works on boolean columns because `True` counts as 1 and `False` counts as 0.

```python
# setup
students = pd.DataFrame(
    {
        'student_id': [101, 102, 103, 104, 105, 106],
        'name': ['Ana', 'Ben', 'Cleo', 'Darius', 'Eli', 'Fern'],
        'group': ['red', 'blue', 'red', 'blue', 'red', 'blue'],
    }
)

scores = pd.DataFrame(
    {
        'student_id': [101, 102, 103, 104, 105, 106],
        'exam1': [92, 75, 68, 84, 59, 88],
        'exam2': [94, 82, 72, 80, 65, 91],
    }
)

# solution
def solution3_2(students: pd.DataFrame, scores: pd.DataFrame) -> tuple[pd.DataFrame, pd.DataFrame]:
    merged = students.merge(scores, on='student_id')
    merged['average_score'] = merged[['exam1', 'exam2']].mean(axis=1)
    merged['passed'] = merged['average_score'] >= 70
    merged['letter_grade'] = 'F'

    merged.loc[merged['average_score'] >= 70, 'letter_grade'] = 'C'
    merged.loc[merged['average_score'] >= 80, 'letter_grade'] = 'B'
    merged.loc[merged['average_score'] >= 90, 'letter_grade'] = 'A'

    summary = merged.groupby('group').agg(
        average_score=('average_score', 'mean'),
        students_passed=('passed', 'sum'),
    ).round(2)

    return merged.round({'average_score': 2}), summary

merged_df, group_summary = solution3_2(students, scores)

print(merged_df, '\n')
print(group_summary)
```

```python
# method chaining solution
def solution3_2_b(students: pd.DataFrame, scores: pd.DataFrame) -> tuple[pd.DataFrame, pd.DataFrame]:
    def score_to_grade(score: float) -> str:
        if score >= 90: return 'A'
        elif score >= 80: return 'B'
        elif score >= 70: return 'C'
        elif score >= 60: return 'D'
        else: return 'E'

    merged: pd.DataFrame = (
        students
        .merge(scores, on='student_id')
        .assign(
            average_score = lambda d: d[['exam1', 'exam2']].mean(axis=1),
            passed = lambda d: d['average_score'] >= 70,
            letter_grade = lambda d: d['average_score'].map(score_to_grade),
        )
    )

    summary = merged.groupby('group').agg(
        average_score=('average_score', 'mean'),
        students_passed=('passed', 'sum'),
    ).round(2)

    return merged.round({'average_score': 2}), summary

merged_df, group_summary = solution3_2(students, scores)

print(merged_df, '\n')
print(group_summary)
```

### The data
- From a delayed discounting experiment with rats.
- Delayed discounting is the decrease of the subjective value of a reward with the delay until recieving the reward. A hamburger now is subjectively more valuble than a hamburger 1 year from now.
- Rats can either choose a small-reward lever that give them 1 food pellet immediately or a large-reward lever that gives them 4 food pellets after a delay.

### The model
We will employ a hyberbolic delay-discounting model (equation 1) to calculate the subjective value of the delayed reward and a Luce's / softmax choice rule (equation 2) to calculate the probability of large-reward choice based on the subjective value.

$$\Delta\text{Reward Value} = \frac{\text{delayed value}}{1 + (k \times \text{delay})} - \text{immediate value}$$

$$
\text{P(delayed choice)} = (1 - \delta) \frac{1}{(1 + \exp({-\beta \times \Delta\text{Reward Value} + \epsilon}))} + \frac{\delta}{2}
$$

Where: \
$\text{delayed value} = 4$ \
$\text{immediate value} = 1$ \
\
$k = \text{behavioral constant}$ \
$\beta = \text{inverse temperature (decision noise)}$ \
$\delta = \text{lapse}$ \
$\epsilon = \text{bias}$
